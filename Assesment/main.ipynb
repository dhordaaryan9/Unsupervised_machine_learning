# 1.  You identify 'Attrition' as the target variable in the IBM HR dataset. What is the difference between a feature and a label in a supervised learning problem? 

**Feature (Input):** The independent variables or attributes used to make predictions (e.g., Age, Monthly Income, OverTime).

**Label (Output):** The dependent variable or the "answer" the model is trying to predict (e.g., Attrition: Yes/No).

# 2. The HR attrition problem predicts a Yes or No outcome. What type of ML problem is this — regression or classification? 

**Classification**: because the output is binary means classified not continous values.

# 3. Your model performs well on training data but poorly on unseen test data. What is this condition called, and what causes it? 

**Conditio**: overfitting 

**Cause**: it happens when model is too complex and memorize the training data instead learning the pattern of data.

# 2. Design a modular Python script to automate the initial data ingestion and feature-label splitting process. 

```python
from sklearn.model_selection import train_test_split
```

```python
def ingest_and_split_data(filepath, target_column, test_size=0.2, random_state=42):

    df = pd.read_csv(filepath)

    X = df.drop(columns=[target_column])
    y = df[target_column]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state, stratify=y)

    print(f"Total records: {len(df)}")
    print(f"Training features shape: {X_train.shape}")
    print(f"Testing features shape: {X_test.shape}")
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = ingest_and_split_data("WA_Fn-UseC_-HR-Employee-Attrition.csv", 'Attrition')
```
